```python
dataset_file = '/content/drive/My Drive/music-Phoebe-Bridgers.wav'
```

```python
loader = ess.MonoLoader(filename= dataset_file)
audio = loader()
```

```python
pitch_extractor = ess.PredominantPitchMelodia(frameSize=2048, hopSize=128)
pitch_values, pitch_confidence = pitch_extractor(audio)
```

```python
from pylab import plot, show, figure, imshow
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15, 6) # set plot sizes to something larger than default

plot(audio)
# for pitch in pitch_values:
    # plt.axvline(x=pitch*20, color='red')

plt.title("Audio waveform and the estimated beat positions")
```

```python
pitch_segmenter = ess.PitchContourSegmentation()
midi_notes = pitch_segmenter(pitch_values, pitch_confidence)
```

```python
from midiutil.MidiFile import MIDIFile
```

```python
def save_midi_notes(notes, filename):
    # Create a MIDIFile object with 1 track
    midi = MIDIFile(1)

    # Set the tempo (optional)
    tempo = 120
    midi.addTempo(0, 0, tempo)

    # Set the instrument (optional)
    instrument = 0  # Piano
    midi.addProgramChange(0, 0, 0, instrument)

    # Add the notes to the MIDI file
    channel = 0
    time = 0
    duration = 1  # Duration of each note in beats
    volume = 100  # Velocity or volume of each note
    for note in notes:
        midi.addNote(channel, 0, note, time, duration, volume)
        time += 1

    # Save the MIDI file
    with open(filename, 'wb') as file:
        midi.writeFile(file)
```

```python
notes = midi_notes
filename = 'output.mid'  
save_midi_notes(notes, filename)
!cp output.mid /content/drive/MyDrive/
```
